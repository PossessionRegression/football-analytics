```python
# Imports packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import preprocessing
import random
```

```python
#Format columns 
def renameCols(features_in_pc):
    cols = features_in_pc.columns.tolist()
    new_cols = ["playing_style_" + str(s) for s in cols]
    features_in_pc.columns = new_cols
    return features_in_pc

# get player details
def getDetails(data):
    principalDf = pd.DataFrame(data = principalComponents)
    min_max_scaler = preprocessing.MinMaxScaler()
    scaled_df = min_max_scaler.fit_transform(principalDf)
    scaled_df = pd.DataFrame(scaled_df)
    
    x = [str(i+1) for i in list(scaled_df.columns.values)]
    s = ["playing_style_" + t for t in x]
    
    scaled_df.columns = s 

    scaled_df["playing_style"] = scaled_df.idxmax(axis=1)
        
    #add player_id to the principalDF 
    data = data.reset_index(drop=True)
    scaled_df["player_id"] = data["player_id"]

    data["playing_style"] = scaled_df["playing_style"]
     
    df = data.groupby(['playing_style'],as_index=False).agg({"player_id":"size"})
    df.columns = ["playing_style", "count"]
    df1 = data.groupby(['playing_style'],as_index=False).mean()
    del df1["player_id"], df1["time_played"], df1["time_attacking"], df1["time_defending"]
    df = df.merge(df1, on = "playing_style", how = "left")

    return df
    
#Classify players by their highest PC
def classifyPlayers(data):
    
    principalDf = pd.DataFrame(data = principalComponents)
    min_max_scaler = preprocessing.MinMaxScaler()
    scaled_df = min_max_scaler.fit_transform(principalDf)
    scaled_df = pd.DataFrame(scaled_df)
    x = [str(i+1) for i in list(scaled_df.columns.values)]
    s = ["playing_style_" + t for t in x]
    scaled_df.columns = s
    scaled_df["playing_style"] = scaled_df.idxmax(axis=1)
    
    #add player_id to the principalDF 
    data = data.reset_index(drop=True)
    scaled_df["player_id"] = data["player_id"]

    return scaled_df

# Reclassify Players with columns names 
def reclassifyPlayers(data, columns):
    
    principalDf = pd.DataFrame(data = principalComponents)
    min_max_scaler = preprocessing.MinMaxScaler()
    scaled_df = min_max_scaler.fit_transform(principalDf)
    scaled_df = pd.DataFrame(scaled_df)
    
    scaled_df.columns = columns

    scaled_df["playing_style"] = scaled_df.idxmax(axis=1)
    
    #add player_id to the principalDF 
    data = data.reset_index(drop=True)
    scaled_df["player_id"] = data["player_id"]
    
    return scaled_df
```
